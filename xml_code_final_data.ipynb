```python
import os 
import pandas as pd
import xml.etree.ElementTree as ET

mrid_list = []
hes_xml_file_name = []
date_list = []

k=1
# Folder Path 
path = r"C:\Users\Ratan Kumar Jha\Desktop\New folder\BP DATA\HES_DP_FILE_DATA"

# Change the directory 
os.chdir(path) 

# Read text File 

def read_text_file(file_path): 
	with open(file_path, 'r') as f: 
		print(f.read()) 

# iterate through all file 
for file in os.listdir(): 
	# Check whether file is in text format or not 
	if file.endswith(".xml"): 
		file_path = f"{path}\{file}"
		print(f"{k} : the  path is : {path}")
		print(f"{k} : the  file is : {file}")

		print(k)
		
		k=k+1
		tree = ET.parse(file_path)
		root = tree.getroot()

		# Define the namespace
		namespace = {'mtr': 'http://iec.ch/TC57/2011/MeterReadings#'}

		for meter_reading in root.findall('.//mtr:MeterReading', namespace):
			mrid = meter_reading.find('mtr:Meter/mtr:mRID', namespace).text
			end = meter_reading.find('mtr:valuesInterval/mtr:end', namespace).text

			mrid_list.append(mrid)
			hes_xml_file_name.append(file)
			date_list.append(end)  # Fixed the variable name here
		
			
data = {
    'Meter ID': mrid_list,
	'hes_xml_file_name' : hes_xml_file_name,
    'date' : date_list
}

hes_dp_Data= pd.DataFrame(data)
```
